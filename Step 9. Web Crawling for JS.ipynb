```python
from bs4 import BeautifulSoup
import sys
import pandas as pd
```

```python
from selenium import webdriver
import time
try:
    driver = webdriver.PhantomJS(executable_path='/usr/local//Cellar/phantomjs/2.1.1/bin/phantomjs')
    driver.get('https://www.morningstar.com/stocks/xshe/000423/quote.html')
    time.sleep(10)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.close()
except:
    print('Do not use VPN !!!')
    print('Do not use VPN !!!')
    print('Do not use VPN !!!')
    print('Err:', sys.exc_info()[0])
```

```python
soupTable = soup.find('table', {'class' : 'report-table ng-isolate-scope'})
```

```python
trList = soupTable.findAll('tr')

data = {}
for tr in trList:
    tds = tr.findAll('td')
    column = [td.getText().strip() for td in tds]
    data[column[0]] = column[1:]

df = pd.DataFrame(data)
df.set_index('Fiscal', inplace=True)
df
```
